```python
plt.figure(figsize=(10,10))
plt.hist(df['Age'])
plt.title("Age distribution")
plt.show()
```

```python
plt.title=('gender distribution')
color=['#c91691','#2573c2']
plt.pie(df.gender.value_counts(),explode=(0.025,0.025),labels=df.gender.value_counts().index,colors=color,autopct='%1.1f%%',startangle=180);
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.countplot(x=df['Your current year of Study'],hue=df['gender'])
plt.title('students studying in particular year')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.countplot(x='Do you have Anxiety?',hue='Do you have Depression?',data=df)
plt.title("Students studyig in particular year");
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.countplot(x='Do you have Anxiety?',hue='gender',data=df)
plt.title("anxiety of male and female");
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.countplot(x='Do you have Depression?',hue='What is your course?',data=df)
plt.legend(title='Hue Category', bbox_to_anchor=(1.5,1), loc='upper right')
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.countplot(x='Do you have Anxiety?',hue='What is your course?',data=df)
plt.legend(title='Hue Category', bbox_to_anchor=(1.5,1), loc='upper right')
plt.show()
```
